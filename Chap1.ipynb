## 1.3 Graphe d'une fonction : exemple

Reprsentez le graphe de la fonction
\begin{equation}

Ajoutez une grille  votre graphe, donnez un nom aux axes et un titre au graphe.<br />

```python
x=np.linspace(-1,1,2001)
```

Le vecteur x ainsi cr reprsente la grille de points en abscisse, grille linairement espace de x=-1 (inclus)  x=1 (inclus), avec 2001 points au total.<br />
Pour voyager dans le vecteur x:

```python
print(x)
print(type(x))
print(len(x))      # Longueur du vecteur
print(x[0])        # premier indice vaut toujours 0
print(x[0:3])      # indices 0  2 (3 non compris!!!)
print(x[:3])       # Idem
print(x[3:])       # Indice 3 compris jusqu' la fin
print(x[-1])       # Dernire valeur de x
print(x[4:12:2])   # De l'lment 4 jusqu' l'lment 12 non compris, par pas de 2
print(x[::2])      # 1 valeur sur 2 sur tout le vecteur
print(x[::-1])     # Vecteur dans l'autre sens - trs utile pour passer de ordre croissant  dcroissant
```

### Autre faon 1 : travailler avec une ***fonction*** python, via "def" 

Privilgier cette faon de faire ou la suivante, qui permettent d'effectuer des oprations sur la fonction ainsi cre, comme des recherches de racine ou d'extrema (voir ci-dessous)

```python
x=np.linspace(-1,1,2001)
def f(z): return 1/(1+25*z**2)
plt.plot(x,f(x),"k")                               
plt.grid()                                   
plt.xlabel("mon axe x")
plt.ylabel("mon super axe y")
plt.title("Mon mga super beau graphe")
#plt.close()  
```

### Autre faon 2 : travailler avec une ***fonction anonyme*** python, via "lambda" 

```python
x=np.linspace(-1,1,2001)
f=lambda z: 1/(1+25*z**2)
plt.plot(x,f(x),"k")                               
plt.grid()                                   
plt.xlabel("mon axe x")
plt.ylabel("mon super axe y")
plt.title("Mon mga super beau graphe")
#plt.close()  
```

## 1.7 Interpolation de donnes : exemple

Le tableau ci-dessous donne la force lectromotrice $E$ en volts d'une pile en fonction de la temprature absolue $T$ en Kelvin. 

|T|290|300|310|320|330|
|-|-|-|-|-|-|
|E|1,15053|1,14950|1,14488|1,14403|1,14325|

Reprsentez les points correspondants aux donnes dans un graphe adquat.<br />
Utilisez les mthodes d'interpolation linaire et spline cubique pour relier les points entre eux,  l'aide de la fonction prdfinie `interp1d` de `scipy.interpolate`. 

```python
T=np.array([290,300,310,320,330])
E=np.array([1.15053,1.14950,1.14488,1.14403,1.14325])
plt.plot(T,E,"b",label="interp linaire")   # Interpolation faite par dfaut graphiquement
plt.plot(T,E,"ro")
f=scint.interp1d(T,E,"cubic")    # f est la fonction d'interpolation par splines cubiques passant les points de coordonnes (T,E)
Tinterp=np.linspace(290,330,100)  
plt.plot(Tinterp,f(Tinterp),"g",label="interp spline cubique")
plt.xlabel("T(K)")
plt.ylabel("E(V)")
plt.grid()
plt.legend()
```
